```python
import os

dir='All Mail';
arr = os.listdir(dir);

for email in arr:
    if (email.endswith(".eml")):
        f = open(dir + '/' + email,'r');
        for x in f: # date, from, and subject must appear in respective order in the eml files
            if (x.startswith("Date")):
                d1 = x;
            if (x.startswith("From")):
                if ("envelop[e]" in x or ("envelopesoundsystem" in x)):
                    break;
                f1 = x;
            if (x.startswith("Subject")): 
                s1 = x;
                print(d1.rstrip() + "/" + f1.rstrip() + "/"  + s1.rstrip());
        f.close();
```
